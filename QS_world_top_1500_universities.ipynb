```python
def clean_rank(rank_value):
    rank_str = str(rank_value)

    if '-' in rank_str:
        return float(rank_str.split('-')[0])

    try:
        return float(rank_str)
    except ValueError:
        return None
```

```python
from scipy import stats

public_citations = df[df['Status'] == 'Public']['Citations per Faculty SCORE'].dropna()
private_citations = df[df['Status'] == 'Private not for Profit']['Citations per Faculty SCORE'].dropna()

t_stat, p_val = stats.ttest_ind(public_citations, private_citations)

print(f"T-статистика: {t_stat:.4f}")
print(f"P-значение (p-value): {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Вывод: Разница значима. Статус вуза влияет на цитируемость.")
else:
    print("Вывод: Разница не значима. Статус вуза не гарантирует более высокую цитируемость.")
```

```python
corr_reputation = df['Employment Outcomes SCORE'].corr(df['Employer Reputation SCORE'])
corr_citations = df['Employment Outcomes SCORE'].corr(df['Citations per Faculty SCORE'])

print(f"Корреляция (Работа + Репутация): {corr_reputation:.4f}")
print(f"Корреляция (Работа + Наука): {corr_citations:.4f}")

if corr_reputation > corr_citations:
    print("\nВывод: Гипотеза подтверждена! Бренд и репутация важнее для карьеры, чем научные успехи.")
else:
    print("\nВывод: Научные достижения вуза ценятся работодателями выше, чем просто репутация.")
```
